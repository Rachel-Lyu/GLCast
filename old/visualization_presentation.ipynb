```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/ruiqil/Documents/PheOpt/PheOpt/PheOpt/JHUcase.csv', index_col=0)
date_list = [datetime.strptime(str(d), '%Y%m%d') for d in df.index]
plt.figure(figsize=(10, 4))
plt.plot(date_list[630:800], df['PA'][630:800], color = 'k', alpha = 0.8, lw = 2)
plt.plot(date_list[630:800], np.repeat(0, 170), color = 'k', alpha = 0.3, lw = 2)
plt.title('COVID Cases in PA')
plt.xticks(rotation = 30);
```

```python
def moving_average_smoother(signal, window_length = 7):
    signal_padded = np.append(np.nan * np.ones(window_length - 1), signal)
    signal_smoothed = (np.convolve(signal_padded, np.ones(window_length, dtype=int), mode="valid")/ window_length)
    return signal_smoothed

df = pd.read_csv('/Users/ruiqil/Documents/PheOpt/PheOpt/PheOpt/JHUcase.csv', index_col=0)
df = df.apply(moving_average_smoother).iloc[6:]
date_list = [datetime.strptime(str(d), '%Y%m%d') for d in df.index]
plt.figure(figsize=(10, 4))
plt.plot(date_list[:794], df['PA'][:794], color = 'k', alpha = 0.8, lw = 2)
plt.plot(date_list[:794], np.repeat(0, 794), color = 'k', alpha = 0.3, lw = 2)
plt.plot([date_list[624], date_list[624]], [0, 10000], color = 'r', alpha = 0.5, lw = 2)
plt.title('COVID Cases in PA')
plt.xticks(rotation = 30);
```
